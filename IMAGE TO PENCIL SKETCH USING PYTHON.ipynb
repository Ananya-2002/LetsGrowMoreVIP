```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path ='tree.jpeg'
```

### Original Image

```python
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(9,5))
plt.imshow(image)
plt.axis('off')
plt.title('original image')
plt.show()
```

### Grey Image

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(9,5))
plt.imshow(grey_image,cmap='gray')
plt.axis('off')
plt.title('Grey image')
plt.show()
```

### Inverted Image

```python
image_invert = cv2.bitwise_not(grey_image)
plt.figure(figsize=(9,5))
plt.imshow(image_invert,cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

### Blur Image

```python
Gsn_image = cv2.GaussianBlur(image_invert,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(9,5))
plt.imshow(Gsn_image,cmap='gray')
plt.axis('off')
plt.title('Blur Image')
plt.show()
```

### Final Sketch

```python
pencil_sketch = cv2.divide(grey_image,255 - Gsn_image, scale = 255)
plt.figure(figsize=(20,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('Final Sketch Image')
plt.show()
```
